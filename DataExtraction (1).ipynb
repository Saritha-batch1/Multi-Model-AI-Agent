```python
import re

blood_sugar_patterns = [
    re.compile(r'blood sugar\s*[:=]?\s*(\d+\.?\d*)'),
    re.compile(r'glucose\s*[:=]?\s*(\d+\.?\d*)'),
    re.compile(r'bs\s*[:=]?\s*(\d+\.?\d*)'),
    re.compile(r'random blood sugar\s*[:=]?\s*(\d+\.?\d*)'),
    re.compile(r'fasting blood sugar\s*[:=]?\s*(\d+\.?\d*)')
]

blood_sugar_values = []

for pattern in blood_sugar_patterns:
    matches = pattern.finditer(cleaned_text_with_numbers)
    for match in matches:
        value = match.group(1)
        blood_sugar_values.append(float(value))

print(f"Extracted {len(blood_sugar_values)} blood sugar values.")
print("First 10 blood sugar values found:", blood_sugar_values[:10])
```

```python
import numpy as np

if blood_sugar_values:
    mean_bs = np.mean(blood_sugar_values)
    median_bs = np.median(blood_sugar_values)
    std_bs = np.std(blood_sugar_values)
    min_bs = np.min(blood_sugar_values)
    max_bs = np.max(blood_sugar_values)

    print(f"\n--- Statistical Summary of Blood Sugar Values ---")
    print(f"Count: {len(blood_sugar_values)}")
    print(f"Mean: {mean_bs:.2f}")
    print(f"Median: {median_bs:.2f}")
    print(f"Standard Deviation: {std_bs:.2f}")
    print(f"Min: {min_bs:.2f}")
    print(f"Max: {max_bs:.2f}")
else:
    print("No blood sugar values were extracted to perform statistical analysis.")
```

```python
import re

death_cause_patterns = [
    re.compile(r'cause of death is (\w+(\s\w+)*)'),
    re.compile(r'died from (\w+(\s\w+)*)'),
    re.compile(r'expired due to (\w+(\s\w+)*)'),
    re.compile(r'(?:cause of death|death cause)\s*[:=]?\s*([a-zA-Z0-9\s,;\-]+)'),
    re.compile(r'(?:died of|died from|death due to)\s*([a-zA-Z0-9\s,;\-]+)')
]

symptom_patterns = [
    re.compile(r'fever'),
    re.compile(r'cough'),
    re.compile(r'nausea'),
    re.compile(r'shortness of breath'),
    re.compile(r'pain'),
    re.compile(r'headache'),
    re.compile(r'fatigue'),
    re.compile(r'dizziness'),
    re.compile(r'vomiting'),
    re.compile(r'diarrhea'),
    re.compile(r'chills'),
    re.compile(r'sore throat'),
    re.compile(r'muscle aches'),
    re.compile(r'loss of taste or smell')
]

print("Defined death_cause_patterns and symptom_patterns.")
```

## Extract Parameters and Assign Variables



```python
death_causes = []
symptoms = []

for pattern in death_cause_patterns:
    matches = pattern.finditer(cleaned_text_with_numbers)
    for match in matches:
        # Assuming the cause of death is in the first capturing group
        death_causes.append(match.group(1).strip())

for pattern in symptom_patterns:
    matches = pattern.finditer(cleaned_text_with_numbers)
    for match in matches:
        # Symptoms are typically the full match of the pattern
        symptoms.append(match.group(0).strip())

print(f"Extracted {len(death_causes)} death causes.")
print(f"First 10 death causes found: {death_causes[:10]}")

print(f"\nExtracted {len(symptoms)} symptoms.")
print(f"First 10 symptoms found: {symptoms[:10]}")
```
