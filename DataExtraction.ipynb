```python
import kagglehub
import os
import pandas as pd

# Download latest version
path = kagglehub.dataset_download("dikshaasinghhh/bajaj")

print("Path to dataset files:", path)

# List the contents of the downloaded directory to see available files
print("Contents of the downloaded directory:")
for root, dirs, files in os.walk(path):
    for name in files:
        print(os.path.join(root, name))


csv_file_path = os.path.join(path, "bajaj.csv")

try:
    df = pd.read_csv(csv_file_path)
    print("Successfully loaded bajaj.csv into a DataFrame.")
    print(df.head())
except FileNotFoundError:
    print(f"Error: bajaj.csv not found at {csv_file_path}. Please check the directory contents listed above to find the correct path.")
except Exception as e:
    print(f"An error occurred while reading the CSV: {e}")

```
