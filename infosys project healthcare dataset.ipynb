```python
import pandas as pd
df = pd.read_csv("healthcare_dataset.csv")
print(df.head())
print(df.info())
```

```python
import re
def clean_text(text):
    if isinstance(text, str):
        text = text.lower()                     # lowercase
        text = re.sub(r'[^a-z0-9\s]', '', text) # remove punctuation
        text = re.sub(r'\s+', ' ', text).strip() # remove extra spaces
        return text
    return text
text_cols = ["Name", "Medical Condition", "Doctor", "Hospital"]
for col in text_cols:
    df[col + "_clean"] = df[col].apply(clean_text)
df.head()
```

```python
condition_counts = df["Medical Condition_clean"].value_counts()
print(condition_counts)
```

```python
df["hospital_word_count"] = df["Hospital_clean"].apply(lambda x: len(x.split()))
df[["Hospital", "hospital_word_count"]].head()
```
